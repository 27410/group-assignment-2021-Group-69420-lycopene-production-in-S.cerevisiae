# Co-factor swap targets

The co-factor balance is important for the fitness of the microorganisms, but in order to optimize the flux of heterologous pathways it can be prudent to swap co-factors in certain reactions. The "CofactorSwapOptimization" algorithm is a variant of Optswap which swaps cofactors in reactions and sees what effect on the fitness function it has. In this case we define the fitness function as lycopene yield on glucose for swapping FAD with NADP since that is the co-factor in our heterologous pathway.

```python
from cameo.strain_design.heuristic.evolutionary.objective_functions import product_yield
from cameo.strain_design.heuristic.evolutionary_based import CofactorSwapOptimization
from cobra.io import read_sbml_model
model = read_sbml_model('models/yeast_lyco.xml')
```

Here we define the co-factor swap algorithm and the parameters in the way described above.

```python
Fluxes=model.optimize()
model.objective = model.reactions.CrtI
py = product_yield(model.reactions.CrtI, model.reactions.r_1714)
swap_optimization = CofactorSwapOptimization(model=model, cofactor_id_swaps=(["FAD_c","FADH2_c"],["NADP_c","NADPH_c"]),objective_function=py)
solution=swap_optimization.run(max_size=10)
```

Below we can see the results of the cofactor swap, it recommends 11 potential swaps that can increase the yield.

```python
solution
```

```python
reactions=solution.data_frame.targets
reactions
```

Below we can see the reactions recommended in more detail.

```python
for t in reactions:
    print("Reactions:")
    for id in t:
        r=model.reactions.get_by_id(id)
        #print(r.name, id+",","Fitness:", solution.data_frame.fitness[id][10])
        s=""
        for rea in range(len(r.reactants)):
            if rea==len(r.reactants)-1:
                s+=r.reactants[rea].name+" ==> "
            else:
                s+=r.reactants[rea].name+" + "
        for p in range(len(r.products)):
            if p==len(r.products)-1:
                s+=r.products[p].name
            else:
                s+=r.products[p].name+" + "
        print(s)
        print()
```

If the co-factors are swapped in the reactions listed above it could lead to an increase in lycopene yield. Since the lycopene pathway used 4 FAD then one would think that co-factors can have a substantial effect on the yield. The fitness can be seen go above 1, which would indicate that you can create more than 1 lycopene from 1 glucose molecule which is impossible.
